# Graficos

## Condición de aplicación | Orden dosis | Grupo Etario

```python
from floweaver import *

group_vac.columns = ['target', 'source', 'middle', 'value']

nodes = {
    'start': ProcessGroup(list(group_vac['source'])),
    'middle': Waypoint(Partition.Simple('middle', group_vac['middle'].unique())),
    'end': ProcessGroup(list(group_vac['target'])),
}

ordering = [['start'], 
            ['middle'],
            ['end']]

bundles = [Bundle('start', 'end', waypoints=['middle'])]

palette = {'Estratégico': '#d8b365', 'Riesgo': '#31a354', 'Salud': '#2c7fb8', '>60': '#756bb1'}

time = Partition.Simple('source', group_vac['source'].unique())
nodes['start'].partition = time
nodes['end'].partition = Partition.Simple('target', group_vac['target'].unique())

sdd = SankeyDefinition(nodes, bundles, ordering, flow_partition=time)

weave(sdd, group_vac,  palette = palette).to_widget()
```

## Condición de aplicación vs Orden dosis

```python
group_vac.columns = ['middle', 'source', 'target', 'value']

nodes = {
    'start': ProcessGroup(list(group_vac['source'])),
    'end': ProcessGroup(list(group_vac['target'])),
}

ordering = [['start'], 
            ['end']]

bundles = [Bundle('start', 'end')]
    
palette = {'Estratégico': '#d8b365', 'Riesgo': '#31a354', 'Salud': '#2c7fb8', '>60': '#756bb1'}
    
time = Partition.Simple('source', group_vac['source'].unique())
nodes['start'].partition = time
nodes['end'].partition = Partition.Simple('target', group_vac['target'].unique())

sdd = SankeyDefinition(nodes, bundles, ordering, flow_partition=time)

weave(sdd, group_vac,  palette = palette).to_widget()
```
